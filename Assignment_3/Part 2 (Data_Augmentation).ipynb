```python
def rotation(image):
#     aa=cv2.imread(image)
    height,width =image.shape[:2]
    
    rotation_matrix=cv2.getRotationMatrix2D((width/2,height/2),270,.7)

    rotated_image=cv2.warpAffine(image,rotation_matrix,(width,height))
    
    return plt.imshow(rotated_image)
    
```
